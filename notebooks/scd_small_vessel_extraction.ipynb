```python
def mask_vessel(images):
    # replace gray pixels into white
    non_vessels = images != 0.5
    images[non_vessels] = 0.0
    return images

device = 'cuda' if torch.cuda.is_available() else 'cpu'
data_path = '/mydata/dlbirhoui/firat/OADAT'
scd_fname_h5 = 'SCD_RawBP.h5'
scd_key = 'labels'

scd_indices = np.arange(0, 35)
scd_obj = dataset.Dataset(
    fname_h5=os.path.join(data_path, scd_fname_h5),
    key=scd_key,
    transforms=lambda x: torch.from_numpy(np.clip(x/np.max(x), a_min=-0.2, a_max=None)),
    inds=scd_indices,
)

# Define grid size
num_images = len(scd_obj)
grid_size = int(np.ceil(np.sqrt(num_images)))  # Create a square grid

# Create a plot
fig, axes = plt.subplots(grid_size, grid_size, figsize=(15, 15))
axes = axes.flatten()

# Iterate over all images in the dataset
for i, scd_image in enumerate(scd_obj):
    # Mask vessel and process the image
    scd_image = mask_vessel(scd_image.squeeze(0).cpu().numpy())

    # Plot the image
    axes[i].imshow(scd_image, cmap='gray')
    axes[i].axis('off')
    axes[i].set_title(f"Index: {scd_indices[i]}")

# Turn off unused axes
for j in range(num_images, len(axes)):
    axes[j].axis('off')

# Adjust layout and show the grid
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import cv2
import torch
import matplotlib.pyplot as plt
from tqdm import tqdm  # For progress tracking

def mask_vessel(images):
    # Replace gray pixels into white
    non_vessels = images != 0.5
    images[non_vessels] = 0.0
    return images

device = 'cuda' if torch.cuda.is_available() else 'cpu'
data_path = '/mydata/dlbirhoui/firat/OADAT'
scd_fname_h5 = 'SCD_RawBP.h5'
scd_key = 'labels'

# Indices to process
scd_indices = np.arange(0, 20000)

# Dataset object
scd_obj = dataset.Dataset(
    fname_h5=os.path.join(data_path, scd_fname_h5),
    key=scd_key,
    transforms=lambda x: torch.from_numpy(np.clip(x / np.max(x), a_min=-0.2, a_max=None)),
    inds=scd_indices,
)

# Variables to store results
images_with_large_blobs = []  # Indices of images with large blobs
large_blob_indices = []       # Indices of those blobs

# Loop through all images
for idx, scd_image in tqdm(enumerate(iter(scd_obj)), total=len(scd_indices)):
    # Convert PyTorch tensor to NumPy
    scd_image_np = mask_vessel(scd_image.squeeze(0).cpu().numpy())

    # Normalize the image to the range [0, 255]
    scd_image_uint8 = (scd_image_np * 255).astype(np.uint8)

    # Apply binary thresholding
    _, binary_image = cv2.threshold(scd_image_uint8, 50, 255, cv2.THRESH_BINARY)

    # Perform connected components analysis
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(binary_image, connectivity=8)

    # Check for blobs larger than 500 pixels
    large_blob_found = False
    thres = 500
    for i in range(1, num_labels):  # Skip label 0 (background)
        area = stats[i, cv2.CC_STAT_AREA]
        if area > thres:
            large_blob_found = True
            break

    # If large blob found, store the index
    if large_blob_found:
        images_with_large_blobs.append((idx, scd_image_np, labels))
        large_blob_indices.append(scd_indices[idx])

# Print results
print(f"Number of images with blobs larger than {thres} pixels: {len(images_with_large_blobs)}")
print(f"Indices of such images: {large_blob_indices}")

# Save the indices if needed
np.save(f'large_blob_indices_{thres}.npy', large_blob_indices)  # Save to a file for later use
```
